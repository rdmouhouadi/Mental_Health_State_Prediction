```python
# Paths

Demographic_path = "C:/Users/Richie MOUHOUADI/Documents/DSTI/ML with Python/Mental_Health_State_Prediction/Data/Clean_Data/Demographic_features_cleaned.csv"
Clinical_path = "C:/Users/Richie MOUHOUADI/Documents/DSTI/ML with Python/Mental_Health_State_Prediction/Data/Clean_Data/Clinical_features_cleaned.csv"
Social_path = "C:/Users/Richie MOUHOUADI/Documents/DSTI/ML with Python/Mental_Health_State_Prediction/Data/Clean_Data/social_features_cleaned.csv"
Insurance_path = "C:/Users/Richie MOUHOUADI/Documents/DSTI/ML with Python/Mental_Health_State_Prediction/Data/Clean_Data/Insurance_features_cleaned.csv"
target_path = "C:/Users/Richie MOUHOUADI/Documents/DSTI/ML with Python/Mental_Health_State_Prediction/Data/Clean_Data/target_column.csv"
```

```python
Demographic = pd.read_csv(Demographic_path)
Clinical = pd.read_csv(Clinical_path)
Social = pd.read_csv(Social_path)
Insurance = pd.read_csv(Insurance_path)
target = pd.read_csv(target_path)

display(Demographic.head())
print()
display(Clinical.head())
print()
display(Social.head())
print()
display(Insurance.head())
print()
target.head()
```

```python
# filtering the columns
Demographic.drop('Unnamed: 0', axis=1, inplace=True)
Clinical.drop('Unnamed: 0', axis=1, inplace=True)
Insurance.drop('Unnamed: 0', axis=1, inplace=True)
target.drop('Unnamed: 0', axis=1, inplace=True)

display(Demographic.head())
print()
display(Clinical.head())
print()
display(Social.head())
print()
display(Insurance.head())
print()
target.head()
```

```python
# Merge all feature DataFrames and add the target column
final_df = pd.concat([Demographic, Clinical, Social, Insurance, target], axis=1)

# Display the resulting DataFrame
final_df.head()
```
